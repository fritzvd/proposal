## Introduction

Agency is an ambiguous term used differently in psychology, sociology and philosophy usually pointing towards to ability to act towards ones intentions effectively. The sociologist perspective is that agency does not happen in a void but is part of a structure ([Giddens 1984](#ref-giddens1984constitution)), that is part of a temporal reality ([Emirbayer and Mische 1998](#ref-emirbayerWhatAgency1998)) or is dependent on its *“ecology”* ([Biesta and Tedder 2007](#ref-biestaAgencyLearningLifecourse2007)).
