# Achieved human agency or loss of choice in Agentic AI for students

Frederik J van Deventer (HAN University of Applied Sciences)

## Introduction

Young people around the globe are growing up in a different digital reality than their parents, where Artificial Intelligence (AI) is playing a big part. Regardless of whether the term of “AI natives” bares truth, there is a new group of students that spend almost 7 hours and 30 minutes online per day ([Kemp 2025](#ref-kemp2025Digital2025Global2025)), on platforms where AI is embedded, e.g. in search algorithms, AI answers, or chat functions often signified with “AI” or ✨. However, this does not mean that these students have an innate skill or that they can see “behind the curtain” of these technologies just as we expected them to in the debunked “digital natives” narrative of the 2000s ([Selwyn 2009](#ref-selwynDigitalNativeMyth2009)).

Over the last decade the question whether Artificial Intelligence (AI) itself exhibits agency has been debated ([Legaspi, He, and Toyoizumi 2019](#ref-legaspiSyntheticAgencySense2019); [Swanepoel 2021](#ref-swanepoelDoesArtificialIntelligence2021)). This distinction between human and technological agency becomes more blurred with the emerging paradigm of Agentic AI (AAI), which is different to previous forms of AI such as Generative AI (GenAI), which we encounter in platforms lik ChatGPT, Claude or Co-pilot ([Naik, Naik, and Naik 2024](#ref-naikChatGPTAllYou2024)), because AAI can operate (semi-)autonomously and pursue complex goals with little human intervention ([Acharya, Kuppan, and Divya 2025](#ref-acharyaAgentic2025)). Both GenAI and AAI largely depend on recent developments in Large Language Models (LLMs). With AAI multiple instances of these GenAI-bots run in the background interfacing each other creating seemingly autonomous and agentic systems, as a evolutionary next step from the GenAI chat-platforms widely in use among students ([Silvennoinen, Aksovaara, and Alanko-Turunen 2025](#ref-silvennoinenStudentsUsageGenAI2025); [Dai 2025](#ref-daiWhyStudentsUse2025)).

This questions the notion of human agency ([Anderson and Rainie 2023](#ref-andersonFutureHumanAgency2023)), as systems now also seem to (1) act with (2) intention and a certain kind of (3) reasoning, three aspects often associated with agency ([Anscombe 1956](#ref-anscombe1956intention); [Davidson 1978](#ref-davidsonIntending1978)). The chordal triad of Emirbayer and Mische ([1998](#ref-emirbayerWhatAgency1998)) suggests placing agency within not just a structure ([Giddens 1984](#ref-giddens1984constitution)) but in what Mead ([1932](#ref-mead1932philosophy)) calls the *temporal* dimension of sociality, using past, present and imagined futures as 3 dimensions on which agency is performed: *iterational*, *practical-evaluative* and *prospective*. Most of these theories are individual and describe the action of an agency up to a certain point: that of acting itself and the effect it causes. Biesta and Tedder ([2007](#ref-biestaAgencyLearningLifecourse2007)) engages with the chordal triad of agency by focussing on the notion of *agency achieved* in an *“ecological understanding”* of agency ([Biesta and Tedder, n.d.](#ref-biestaHowAgencyPossible)); it is here that the effect of action becomes realized and experienced.

The *iterational* aspect is exposed in structure and collective beliefs or ideology expressed in instruments such as policies, or punitive action or attitudes of individuals. Beliefs about “inevitability of technology” and the cult of innovation ([Winner 2018](#ref-winnerCultInnovationIts2018)) pervade, while at the same time mistrust and weariness tries to keeps AI out of universities ([Guest et al. 2025](#ref-guestUncriticalAdoptionAI2025)). Technology is never just neutral nor is it inherently good or bad ([Morrow 2014](#ref-morrowWhenTechnologiesMakes2014); [Heyndels 2023](#ref-heyndelsTechnologyNeutrality2023)) it is be assistive to our human needs ([Arora 2024, 32](#ref-arora2024pessimism)). Regulation and policy fueled by ideology (both benefiting the few or the many) has been the driver for the direction this advancement would take us ([Johnson and Acemoglu 2023, 57](#ref-johnsonPowerProgressOur2023)).

Students navigate these punitive faculties, policies and teacher narratives as ecology in which they need to realize goals, their agency changing from situation to situation. Are they seeing their volitions realized as *achieved agency*?

## Knowledge Gap

Most literature focusing on AI is not (yet) mentioning agentic AI, but using GenAI, which is closely related as precursor to Agentic AI paradigm that also utilizes LLMs. So a number of the studies mentioned below are chosen from that context.

In the course of history, human development has been argued to increase where agency and freedom increase ([Prados de la Escosura 2022](#ref-PradosdelaEscosura_2022); [Sen 1999](#ref-senDevelopmentFreedom1999)). Agency is however a bloated term that can mean a number of things in different domains, such as psychology where it focuses on individual human agency ([Bandura 2001](#ref-bandura2001social)) or sociology where it focuses on the structures in which an actor operates ([Giddens 1984](#ref-giddens1984constitution)). Agency is a term closely related to and with overlaps to terms such as autonomy and freedom. To be clear in this context, “agency” means for an actor to be free to act true to that same actor’s intentions, considering their circumstance (structure),their beliefs (iterational aspect of Emirbayer and Mische ([1998](#ref-emirbayerWhatAgency1998))).

This in itself is impossible to measure cleanly, being dependent on so many factors. But some attempts have been made to measure how people perceive their *sense of agency* ([Tapal, Oren, and Eitam 2017](#ref-tapalsenseofagencyscale2017)) or how Trust in Automation (TiA) influences there autonomy ([Kohn et al. 2021](#ref-kohnMeasurementTrustAutomation2021)). There are also different models that propose “measures of agency” ([Grünbaum and Christensen 2020](#ref-grunbaumMeasuresAgency2020)) or how concepts like autonomy and ability contribute to agency or measuring proxies for agency ([Alkire 2008](#ref-alkire2008concepts)).

Agentic AI and AI goals in general, largely seem to focus on increasing abilities for models or replacing tedious tasks performed by humans. While research has been done about future projections of human agency in conjunction with AI in decision making ([Pew Research Center, Anderson, and Rainie 2023](#ref-pewFutureHuman2023)) and in education ([Mouta, Pinto-Llorente, and Torrecilla-Sánchez 2025](#ref-moutaWhereAgencyMoving2025)) it does not deal with how to increase human agency for humans now. Or what the effect is of Agentic AI on agency or Sense of Agency (SoA).

The way Agentic AI or GenAI is influencing agency in students seems not to have been studied, apart from the influence it has on critical thinking ([Suriano et al. 2025](#ref-surianoStudentInteractionChatGPT2025)), and why and how they use it for studying ([Dai 2025](#ref-daiWhyStudentsUse2025); [Silvennoinen, Aksovaara, and Alanko-Turunen 2025](#ref-silvennoinenStudentsUsageGenAI2025); [Suonpää, Heikkilä, and Dimkar 2024](#ref-suonpaaStudentsPerceptionsGenerative2024)). The Human AI - Interaction framework suggested by Sundar ([2020](#ref-sundarRiseMachineAgency2020)) incorporates agency, but mostly highlights interesting avenues to pursue rather than diving into topics such as: human and AI collaboration, the seeming either/or thinking of machine vs human agency; agency is seen as a “mediating variable” in the machine/human interaction ([Sundar 2020](#ref-sundarRiseMachineAgency2020)).

One area of study related to increasing human agency surrounding AI is *AI Literacies*. Learning about AI systems explains how they perform, this demystifies and increases people’s insight in these systems ([Pinski and Benlian 2024](#ref-pinskiAILiteracyUsers2024)) which helps to reduce anthropomorphism ([Druga and Ko 2021](#ref-drugaHowChildrensPerceptions2021)). Anthropomorphism is one of the things that can lead to false perceptions and misunderstanding of AI abilities([Barrow 2024](#ref-barrowAnthropomorphismAIHype2024)). AI Literacies among students has not been studied in relation to neither human agency, their sense of agency, nor other similar approaches such as a capabilities approach.

In goal attainment and AI assisted coaching, there is a real opportunity for an AI to participate in the process this then is suggested to increase agency ([Plotkina and Sri Ramalu 2024](#ref-plotkinaUnearthingAICoaching2024)). A review collecting these kinds of studies, that are influential to agency could not be found.

Then temporal aspect of agency as suggested in the above mentioned *chordal triad of agency* seems not have been used at all towards agentic AI or GenAI.

## Contribution to scientific discourse

The threats and opportunities for the achieved agency of students are unclear in the context of Agentic AI. First of all collecting work on how AAI and GenAI are shaping agency needs to be reviewed. Secondly I want to research the *iterational* dimension ([Emirbayer and Mische 1998, 975](#ref-emirbayerWhatAgency1998)) by uncovering what beliefs or ideology is encapsulated in teacher narratives, rules and policy surrounding (A)AI. Thirdly I want to Lastly I will investigate how students enact their intentions within their temporal-relational context of action and how their achieving of agency feedbacks into shaping their context, by looking at how students react to changing policies. This means looking into the *prospective* dimension of the chordal triad ([Emirbayer and Mische 1998, 984](#ref-emirbayerWhatAgency1998)), where goals or aspirations are set and then actualized through the *practical-evaluative* dimension.

## Research Questions

In what way is *agency achieved* impacted for students as a result of changing circumstances with the rise of agentic AI?

### Sub-questions

1.  How can we categorize known effects of GenAI and Agentic AI on agency systematically?
2.  How are prevalent beliefs disclosed in the ecology of achieved agency changing policy and teacher narratives because of AAI?
3.  In what ways are students performing day-to-day tasks, by either conforming or subverting rules and restrictions?

### RQ1: Systematic Literature Review

To find out what studies related to agency a systematic literature will be conducted that searches for studies that relate to agency, such as studies that focus on empowerment ([Kong, Zhu, and Yang 2025](#ref-kongDevelopingValidatingScale2025)). According to the systems of Grünbaum and Christensen ([2020](#ref-grunbaumMeasuresAgency2020)) and Alkire ([2008](#ref-alkire2008concepts)) when studies are performed on proxies of agency, or related concepts, they can be categorized and labeled in relation to AI or more specifically to Agentic AI if those are available. Furthermore the way agency is approached is important to delineat, whether it falls under sociology, social psychology or philosophy

### RQ2: Critical Discourse Analyis

Through *critical discourse analyis* (CDA) I will look at what kind of language is constructing narratives, describes relations of power and how it creates or reinforces myths ([Blommaert and Bulcaen 2000](#ref-blommaertCriticalDiscourseAnalysis2000)).

CDA tries to find patterns in power dynamics and social effects in discourse. In this analysis we can take the texts further and analyze context and intertextuality to challenge social inequalities. I will focus mainly on step two: *Identify obstacles to addressing the social wrong* of this process which consists of three stages ([Fairclough 2013](#ref-fairclough2013critical)):

1.  Analyse dialectical relations
2.  Select texts
3.  Analyse texts interdiscursively and linguistically

### RQ3: Mixed-method

*Thematic analysis* will be used as Naeem et al. ([2023](#ref-naeemStepbyStepProcessThematic2023)) and Braun and Clarke ([2021](#ref-braun2021thematic)) describe it to find common themes and topics in personal qualitative interview with students. Thematic analysis is a more subjective approach to data, where language is central. It requires a *coding* process that uses semantic and latent codes to categorize statements and find *themes* and *subthemes* of meaning in texts.
