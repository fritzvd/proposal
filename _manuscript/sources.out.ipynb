## Key Statistics and Findings

### Workforce Trends

-   **Tokyo University Study**: 97% of students reported using GPT for coursework, indicating a new generation of aid workers familiar with AI tools

### Applications

-   **Advocacy and Communication**: ChatGPT used for drafting social media posts, media releases, and articles
-   **Data Analysis**: AI enables analysis of vast datasets to predict trends and inform humanitarian response
-   **Predictive Analytics**: Shift from reactive to anticipatory humanitarian action
-   **Legal Analysis**: Bespoke generative AI instances to query refugee law and understand protection needs

### Challenges and Risks

-   **Data Privacy**: ChatGPT privacy statement indicates staff read and record information to improve algorithms
    -   Risk: Humanitarians must avoid inputting sensitive operational data or personally identifiable information
-   **Algorithmic Bias**: Poor data quality and biased algorithms can lead to unfair outcomes
-   **Equity Concerns**: Potential to exacerbate vulnerabilities and inequalities
