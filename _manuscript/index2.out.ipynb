## Introduction

New ways to use Generative Artifical Intelligence (genAI) for increasing productivity \[\], criminal behaviour ([Ferrara 2024](#ref-ferraraGenAIHumanityNefarious2024)), berievement ([Su 2025](#ref-suRightRealityHuman2025)) are coming at us fast. GenAI is rapidly reshaping how we work ([Joshi 2025](#ref-joshi2025transformative)), how we sleep \[\], and how learn ([Perifanou and Economides 2025](#ref-perifanouCollaborativeUsesGenAI2025)). Life after AI comes at a tremendous physical cost for our planet ([Crawford 2021](#ref-crawford2021atlas)). The physical extraction of metals for chips, the power consumption to drive the machines and then again for cooling is outgrowing the planned construction of the power grid at a breakneck speed ([Lin et al. 2024](#ref-linExplodingAIPower2024)).

The impact will be most felt by people where labour and land are cheap.

Worldwide the largest population group is estimated to be youth between 15-25 . Rural youth in South Africa is suffering from high numbers of emotinoal distress ([Mngoma et al. 2021](#ref-mngomaDistressDesperationDespair2021)) despite the general relation U-shape between age and happiness ([Graham and Ruiz Pozuelo 2017](#ref-grahamHappinessStressAge2017); [Blanchflower 2021](#ref-blanchflowerHappinessUshapedEverywhere2021)).
